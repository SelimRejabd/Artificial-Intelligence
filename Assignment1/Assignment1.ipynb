```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-10, 10, 0.1)
```

## Plotting y = w * x + b

```python
w = 2
b = 3
y1 = w * x + b
plt.figure(figsize=(12, 6))

plt.plot(x, y1, label=f'y={w}x+{b}', color='green', linewidth=2)
plt.title(f'Plot of y = {w}x + {b}')
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.grid()
plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=2)

plt.tight_layout()
plt.show()
```

## Plotting y = x ** 2

```python

y2 = x ** 2
plt.figure(figsize=(12, 6))

plt.plot(x, y2, label=f'y=x^2', color='green', linewidth=2)
plt.title('Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.grid()

plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=2)

plt.tight_layout()
plt.show()
```

## Plotting y = 1/ (1 + e^(-u))

```python
y3 = 1 / (1 + np.exp(-x))
plt.figure(figsize=(12, 6))

plt.plot(x, y3, color='green', linewidth=2)
plt.title(f'Plot of y =  1/ (1 + e^(-u))')
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.grid()

plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=2)

plt.tight_layout()
plt.show()
```

## Plotting y = (e^x - e^(-x)) / (e^x - e^(-x))

```python

y4 = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
plt.figure(figsize=(12, 6))

plt.plot(x, y4, color='green', linewidth=2)
plt.title('Plot of y =  (e^x - e^(-x)) / (e^x - e^(-x))')
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.grid()

plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=2)

plt.tight_layout()
plt.show()
```
